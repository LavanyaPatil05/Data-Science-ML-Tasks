```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # True labels (not used in clustering)

# Initialize KMeans with 3 clusters (since there are 3 species of Iris in the dataset)
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
kmeans.fit(X)

# Get cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Use PCA to reduce the dataset to 2 dimensions for visualization
pca = PCA(2)
X_pca = pca.fit_transform(X)

# Plot the clusters and the cluster centers
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(pca.transform(centers)[:, 0], pca.transform(centers)[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Centroids
plt.title('K-Means Clustering on Iris Dataset (PCA reduced)')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # True labels (not used in clustering)

# Initialize KMeans with 3 clusters (since there are 3 species of Iris in the dataset)
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
kmeans.fit(X)

# Get cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Use PCA to reduce the dataset to 2 dimensions for visualization
pca = PCA(2)
X_pca = pca.fit_transform(X)
centers_pca = pca.transform(centers)  # Transform cluster centers to PCA space

# Plot the clusters and the cluster centers
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Centroids

# Add labels for each cluster center
for i, (x, y) in enumerate(centers_pca):
    plt.text(x, y, f'Cluster {i+1}', fontsize=12, weight='bold', ha='center', va='center', color='black')

# Customize plot appearance
plt.title('K-Means Clustering on Iris Dataset (PCA reduced)')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features

# List to store inertia values
inertia = []

# Test different numbers of clusters (from 1 to 10)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)  # Store the inertia (sum of squared distances)

# Plotting the elbow graph
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Sum of squared distances)')
plt.xticks(range(1, 11))  # Ensure x-ticks show each value of k
plt.grid(True)
plt.show()
```
