```python
import matplotlib.pyplot as plt

# Create the data vector
v = [19, 23, 11, 5, 16, 21, 32, 14, 19, 27, 39]

# Line Plot
plt.figure(figsize=(10,5))
plt.plot(v, marker='o', linestyle='-', color='blue')
plt.title('Line Plot of v')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()

# Bar Plot
plt.figure(figsize=(10,5))
plt.bar(range(len(v)), v, color='green')
plt.title('Bar Plot of v')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the data vector
v = [19, 23, 11, 5, 16, 21, 32, 14, 19, 27, 39]

# Histogram
plt.figure(figsize=(10,5))
plt.hist(v, bins=5, color='orange', edgecolor='black')
plt.title('Histogram of v')
plt.xlabel('Value Ranges')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot
plt.figure(figsize=(10,5))
plt.scatter(range(len(v)), v, color='red')
plt.title('Scatter Plot of v')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the data vector
v = [19, 23, 11, 5, 16, 21, 32, 14, 19, 27, 39]

# Box Plot
plt.figure(figsize=(10,5))
plt.boxplot(v)
plt.title('Box Plot of v')
plt.ylabel('Values')
plt.grid(True)
plt.show()

# Pie Chart
labels = [f'Value {i+1}' for i in range(len(v))]  # Labels for each slice
plt.figure(figsize=(7,7))
plt.pie(v, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of v')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Create the data vector
v = [19, 23, 11, 5, 16, 21, 32, 14, 19, 27, 39]

# QQ Plot
plt.figure(figsize=(10,5))
stats.probplot(v, dist="norm", plot=plt)
plt.title('QQ Plot of v')
plt.grid(True)
plt.show()

# Density Plot
plt.figure(figsize=(10,5))
sns.kdeplot(v, shade=True, color='purple')  # Kernel Density Estimation
plt.title('Density Plot of v')
plt.xlabel('Values')
plt.ylabel('Density')
plt.grid(True)
plt.show()

```
